### Drop the unsuitable columns

```python
# # drop the STANDAARDDEVIATIE and %Defintieve aansl (NRM) columns
Etrain = Etrain.drop(['STANDAARDDEVIATIE', '%Defintieve aansl (NRM)','Type'], axis=1)
Etest = Etest.drop(['STANDAARDDEVIATIE', '%Defintieve aansl (NRM)','Type'], axis=1)
Etrain.head()
```

```python
print('Unique net manager for Etrain :\n',Etrain['net_manager'].value_counts())
print('Unique net manager for Etest :\n',Etest['net_manager'].value_counts())
print('Unique purchase area for Etrain :\n',Etrain['purchase_area'].value_counts())
print('Unique purchase area for Etest :\n',Etest['purchase_area'].value_counts())
```

```python
print("White space in the train dataset")
def checkExtraWhiteSpace(ColName):
    Variable = [] 
    for anwser in Etrain[ColName].unique():
        for Com_Val in str(anwser).split(";"):
            if (Com_Val in Variable): 
                break
            words = Com_Val.split(" ")
            if ("" in words):
                print("Column '{}', variable '{}' contains extra whitespace.".format(ColName, Com_Val)
                )
                Variable.append(Com_Val)
                break
for column in Etrain.columns:
    if (str(Etrain[column].dtypes) == "object"):
        checkExtraWhiteSpace(column)
print("for testing dataset")
def checkExtraWhiteSpace(ColName):
    Variable = [] 
    for anwser in Etest[ColName].unique():
        for Com_Val in str(anwser).split(";"):
            if (Com_Val in Variable): 
                break
            words = Com_Val.split(" ")
            if ("" in words):
                print("Column '{}', variable '{}' contains extra whitespace.".format(ColName, Com_Val)
                )
                Variable.append(Com_Val)
                break
for column in Etest.columns:
    if (str(Etest[column].dtypes) == "object"):
        checkExtraWhiteSpace(column) 
```

```python
# remove the whitespace in the columns
Etrain['street'] = Etrain['street'].str.strip()
Etest['street'] = Etest['street'].str.strip()
```

```python
# apply uppercase to all object columns since they are mostly unique
def uppercaseAllObjectColumns(dataframe):
    for column in dataframe.columns:
        if (str(dataframe[column].dtypes) == "object"):
            dataframe[column] = dataframe[column].str.upper()
            
uppercaseAllObjectColumns(Etrain)
uppercaseAllObjectColumns(Etest)
Etrain.head(3)
```
